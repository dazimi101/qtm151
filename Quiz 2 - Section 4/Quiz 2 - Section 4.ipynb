(d) Create a function with four inputs $f(y,x,b0,b1)$

<font size = "3">

$\qquad e2 = (y - b_0 - b_1 * x)^2$

- The function should return the sum of e2 entries
- This is the residual sum of squares (rss)
- Call the function "fn_rss"
- Let $y$ be the column "Overall" (overall score)
- Let $x$ be the column "Age" 
- Does <br>
$\qquad (b_0 = 50, b_1 = -6 )$ or <br>
$\qquad (b_0 = 50, b_1 = 0.6 )$ or <br>
$\qquad (b_0 = 50, b_1 = 6)$ <br>
produce the lowest value of rss?


````python
def fn_rss(y, x, b0, b1):
    e2 = (y - b0 - (b1 * x)) ** 2
    return np.sum(e2)

rss_1 = fn_rss(fifa23["Overall"], fifa23["Age"], 50, -6)
rss_2 = fn_rss(fifa23["Overall"], fifa23["Age"], 50, .6)
rss_3 = fn_rss(fifa23["Overall"], fifa23["Age"], 50, 6)
print("Using b0 = 50, b1 = -6 produces rss of " + str(rss_1) + ", using b0 = 50, b1 = .6 produces rss of " + str(rss_2) +
      ", and using b0 = 50, b1 = 6 produces rss of " + str(rss_3) + ". b0 = 50, b1 = .6 produces the lowest rss value.")
    
````

(e) Split a dataset into subsets

<font size = "3">

- Create ``` list_status = [1,2]```
- Create ``` list_prob = [0.5, 0.5]```
- Create a new variable $n$ with the number of rows of the dataset
- Use 
```python
    np.random.choice(list_status,
                     size = n,
                     p = list_prob)
```
$\qquad$ to create a new column called "status" <br>
$\qquad$ which randomly assigns each row to "1" or "2" <br>
$\qquad$ with equal probability
- Use .query() to subset all observations with status == 1.<br>
store it in a new dataset called "data1"
- Use .query() to subset all observations with status == 2.<br>
store it in a new dataset called "data2"



````python
# Write your own code

list_status = [1, 2]
list_prob = [.5, .5]
n = len(fifa23)

fifa23["status"] = np.random.choice(list_status, size = n, p = list_prob)

data1 = fifa23.query("status == 1")
data2 = fifa23.query("status == 2")

display(data1)
display(data2)


````

(f) Create two overlapping histogram plots

<font size = "3">

- This function requires the two datasets you created in (e)
- Display the distribution of "Potential" for "data1" and "data2"
- Use the "alpha" option to make the graphs semitransparent
- You will need to add a legend, label the axes, and the title

Note: The goal of this question is to illustrate that random <br>
assignment produces very similar distributions between two groups

````python
# Write your own code
list_datasets = [data1, data2]
list_datasets_names = ["Status = 1", "Status = 2"]

for dataset in list_datasets:
    plt.hist(x = dataset["Potential"], alpha = .5)

plt.legend(labels = list_datasets_names)
plt.title("Distribution of Player Potential based on Status (1 or 2)")
plt.xlabel("Potential")
plt.ylabel("Frequency")
plt.show()
````
